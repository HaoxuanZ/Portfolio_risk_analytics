```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
df_1 = yf.download(['^GSPC','^VIX'],start = '2007-01-01',end='2020-03-31')['Close']
df_1
```

```python
df_1['GSPC_rets']= df_1['^GSPC'].pct_change()
df_1['VIX_rets']=df_1['^VIX'].pct_change()
df_1
```

```python
df_1['Real_21std'] = np.sqrt(252)*df_1['GSPC_rets'].rolling(window=21,min_periods=1).std()
df_1
```

```python
df_1['VIX_1m_back'] = df_1['^VIX'].shift(21)
df_1 = df_1.dropna()
df_1
```

```python
df_1['Equity_Allocation'] = np.where(100*df_1['Real_21std']<=df_1['VIX_1m_back']+2,1,0.4)
df_1
```

```python
df_1['Volatility_Allocation'] = 1 - df_1['Equity_Allocation'] 
df_1
```

```python
df_1['Portfolio_Return'] = df_1['Equity_Allocation']*df_1['GSPC_rets'] +df_1['Volatility_Allocation']*df_1['VIX_rets']
```

```python
df_2 = df_1[['Portfolio_Return']]
df_2['Portfolio_cum_rets'] = (1+df_1['Portfolio_Return']).cumprod()-1
df_2['GSPC_cum_rets'] = (1 + df_1['GSPC_rets']).cumprod() -1
df_2
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create a figure
fig = go.Figure()

# Add trace, x axis, y axis, mode = lines
fig.add_trace(go.Scatter(x=df_2.index, y=df_2['Portfolio_cum_rets'], mode='lines',line=dict(color='royalblue',width=2), name='Portfolio'))
fig.add_trace(go.Scatter(x=df_2.index, y=df_2['GSPC_cum_rets'], mode='lines',line=dict(color='red',width=2), name='SPY'))


# Update Layout
fig.update_layout(
    title={'text':'Cumulative Performance',
           'y':0.9,
        'x':0.5,
          'xanchor': 'center',
             'yanchor': 'top' },
    
    xaxis=dict(
        title='',
        #ticksuffix="%",
        #tickprefix=''
    ),
    
    yaxis=dict(
        title='Returns',
         ticksuffix="%"
    ),
    
 
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)',
)





```

```python
n = len(df_2)
Port_Ann_Ret = 100* ( (1+df_2["Portfolio_cum_rets"][-1])**(252/n) - 1 )
print('Portfolio Annualized Return: ',np.round(Port_Ann_Ret,2),'%')


Port_ann_vol = np.sqrt(252)*df_2['Portfolio_Return'].std()
print("Portfolio Annualized Volatility: ", np.round(100*Port_ann_vol,2),"%")
```

```python
n = len(df_1)
SPY_Ann_Ret = 100* ( (1+df_1["GSPC_cum_rets"][-1])**(252/n) - 1 )
SPY_ann_vol = np.sqrt(252)*df_1['GSPC_rets'].std()


print('\n\nSPY Annualized Return: ',np.round(SPY_Ann_Ret,2),'%')
print("SPY Annualized Volatility: ", np.round(100*SPY_ann_vol,2),"%")
```

```python
sharpe_ratio = (Port_Ann_Ret/100)/Port_ann_vol
print("Strategy SR: ", np.round(sharpe_ratio,2))


# Sharpe Ratio SPY
sharpe_ratio_SPY = (SPY_Ann_Ret/100)/SPY_ann_vol
print("SPY SR: ",np.round(sharpe_ratio_SPY,2))
```
